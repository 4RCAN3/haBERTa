```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
device
```

```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

model = T5ForConditionalGeneration.from_pretrained("sana-ngu/HaT5").to(device)
tokenizer = T5Tokenizer.from_pretrained("t5-base")
tokenizer.pad_token = tokenizer.eos_token
input_ids = tokenizer(list(df['text']), padding=True, truncation=True, return_tensors='pt').input_ids.to(device)
```

```python
input_ids[0].shape
```

```python
preds = []
for text in list(df['text']):
  input_ids = tokenizer(text, padding=True, truncation=True, return_tensors='pt').input_ids.to(device)
  op = model.generate(input_ids)
  pred = tokenizer.decode(op[0], skip_special_tokens=True)
  print(pred)
  preds.append(pred)
```

```python
set(preds)
```

```python
res = [0 if pred == '<pad> 1' else 1 for pred in preds]
```

```python
torch.cuda.empty_cache()
```

```python
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score
print('roc: ', roc_auc_score(df['label'], res))
print('f1: ', f1_score(df['label'], res, average = 'macro'))
print('accuracy: ', accuracy_score(df['label'], res))
```
