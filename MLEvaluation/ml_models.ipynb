```python
sbert_model = 'all-mpnet-base-v2'
```

```python
import pandas as pd

df = pd.read_csv('non_skewed_data.csv')
documents = df['tweet']
```

```python
documents = list(documents)
```

```python
from sentence_transformers import SentenceTransformer
embedding_model = sbert_model
model = SentenceTransformer(embedding_model)
embeds = model.encode(documents)
```

```python
embeds.shape
```

```python
df['embeds'] = embeds.tolist()
```

```python
df
```

```python
df['label'] = df['subtask_a'].apply(lambda x: 1 if x == 'OFF' else 0)
```

```python
df = df.iloc[:, 5:]
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(
    df['embeds'].tolist(), df['label'], test_size=0.2, random_state=2
)
```

```python
svm_classifier = SVC(kernel='linear', C=1.0)  # You can experiment with different kernel functions and C values
svm_classifier.fit(train_data, train_labels)
```

```python
svm_predictions = svm_classifier.predict(test_data)

# Evaluate the model
'''accuracy = accuracy_score(test_labels, predictions)
conf_matrix = confusion_matrix(test_labels, predictions)
classification_rep = classification_report(test_labels, predictions)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", classification_rep)'''
```

```python
from sklearn.naive_bayes import GaussianNB

nb_classifier = GaussianNB()
nb_classifier.fit(train_data, train_labels)

# Make predictions on the test set
nb_predictions = nb_classifier.predict(test_data)

# Evaluate the model
'''accuracy = accuracy_score(test_labels, predictions)
conf_matrix = confusion_matrix(test_labels, predictions)
classification_rep = classification_report(test_labels, predictions)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", classification_rep)'''
```

```python
svm_preds = svm_classifier.predict(test_data)
knn_preds = knn_classifier.predict(test_data)
nm_preds = nb_classifier.predict(test_data)
```

```python
import numpy as np
preds = np.array([svm_preds, knn_preds, nm_preds]).T
preds = preds.tolist()
```

```python
ensemble = []
for result in preds:
  count_0 = result.count(0)
  count_1 = result.count(1)
  ensemble.append(1 if (count_1 > count_0) else 0)

ensemble
```

# Ensemble

```python
from sklearn.metrics import f1_score, roc_auc_score, accuracy_score
print('roc: ', roc_auc_score(test_labels, ensemble))
print('f1: ', f1_score(test_labels, ensemble, average = 'macro'))
print('accuracy: ', accuracy_score(test_labels, ensemble))
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

def plot_roc_curve(y_true, y_scores, model_name):
    fpr, tpr, _ = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    plt.plot(fpr, tpr, label=f'{model_name} (AUC = {roc_auc:.2f})')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
```

```python
from sklearn.metrics import precision_recall_curve, auc

def plot_precision_recall_curve(y_true, y_scores, model_name):
    precision, recall, _ = precision_recall_curve(y_true, y_scores)
    pr_auc = auc(recall, precision)
    plt.plot(recall, precision, label=f'{model_name} (AUC = {pr_auc:.2f})')
```

```python
models = [svm_preds, knn_preds, nm_preds, ensemble]
plt.figure(figsize=(8, 6))
plot_roc_curve(test_labels, svm_preds, 'SVM')
plot_roc_curve(test_labels, knn_preds, 'KNN')
plot_roc_curve(test_labels, nm_preds, 'Gaussian Naive Bayes')
plot_roc_curve(test_labels, ensemble, 'Ensemble')

plt.plot([0, 1], [0, 1], linestyle='--', color='r', label='Random Guessing')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plot_precision_recall_curve(test_labels, svm_preds, 'SVM')
plot_precision_recall_curve(test_labels, knn_preds, 'KNN')
plot_precision_recall_curve(test_labels, nm_preds, 'Gaussian Naive Bayes')
plot_precision_recall_curve(test_labels, ensemble, 'Ensemble')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

res = [svm_preds, knn_preds, nm_preds, ensemble]
names = ['SVM', 'KNN', 'Gaussian Naive Bayes', 'Ensemble']

plt.figure(figsize=(15, 10))
for i, y_pred in enumerate(res):
    plt.subplot(2, 3, i + 1)
    cm = confusion_matrix(test_labels, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)
    plt.title(names[i])
    plt.xlabel('Predicted')
    plt.ylabel('True')

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import accuracy_score, f1_score
import numpy as np

accuracy_scores = [accuracy_score(test_labels, y_pred) for y_pred in res]
f1_scores = [f1_score(test_labels, y_pred, average='macro') for y_pred in res]

plt.figure(figsize=(10, 5))
x = np.arange(len(names))
width = 0.35
bars1 = plt.bar(x - width/2, accuracy_scores, width, label='Accuracy')
bars2 = plt.bar(x + width/2, f1_scores, width, label='F1 (Macro)')
plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Accuracy and F1 Score Comparison for Machine Learning Models')
plt.xticks(x, names)
plt.legend(loc = 'lower right')

# Add text labels for values
for bars in [bars1, bars2]:
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.2f}', ha='center', va='bottom')

plt.show()
```
