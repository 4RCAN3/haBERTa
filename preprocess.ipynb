```python
  import nltk
  from nltk.corpus import stopwords
  from nltk.tokenize import word_tokenize

  nltk.download('stopwords')
  nltk.download('punkt')

  def remove_stopwords(tweet):
      stop_words = set(stopwords.words('english'))
      words = word_tokenize(tweet)
      filtered_tweet = [word for word in words if word.lower() not in stop_words]
      return ' '.join(filtered_tweet)
```

```python
def remove_punctuation(tweet):
    tagged_words = re.findall(r'<[^>]+>|[\w]+', tweet)
    cleaned_words = [re.sub(r'[^\w\s]', '', word) if not word.startswith('<') else word for word in tagged_words]
    processed_tweet = ' '.join(cleaned_words)

    return processed_tweet
```

```python
import re
import emoji

def preprocess(tweet):
    tweet = re.sub(r'@([a-zA-Z0-9_]+)', r'<user>', tweet)
    tewet = tweet.replace('URL', '<url>')
    tweet = re.sub(r'https?://\S+', r'<url>', tweet)
    tweet = emoji.demojize(tweet, delimiters=("<emoji:", ">"))
    tweet = tweet.lower()
    tweet = remove_punctuation(tweet)
    tweet = remove_stopwords(tweet)

    return tweet
```

```python
df['text'] = df['text'].apply(lambda x: preprocess(x))
df
```
